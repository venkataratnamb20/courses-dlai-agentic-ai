# Systematic Literature Review

## Steps

1. Identify Basic characteristics
- Field of Study
- Relevent Keywords and create phrases, with keywords, to search.
- Databases to search
2. Search for phrases and collect the database of references
- Tools: Web Search, IEEE, xArchiv, google Scholar, etc
3. Process the database of references and shortlist the most relevent references [top 100]
- Process
- Download
- Organize
4. Process downloaded references
- Make a summary of each reference
- Make a data mapping table
- Discard duplicate and irrelevent references
5. Write a first draft with data mapping table and summuries
6. Identify improvements in each section/topic
- Identify improvements
- Get more keywords if necessary
- Create and/or refine phrases for refined search.
7. Research
- Search again with new phrases if references are not enough.
8. Write final Draft.



## Definitions and Criteria

1. **Field of Study**
- Technology
- Medical
- Finance
- Geography
- History
- Sociology
- Political
- Education
- Culture & Religion
2. **Relevence**
- contains more keywords or phrases
- recently published
- Mostly read
- Mostly referenced
- Author credibility
- University/organization credibility
3. **Author Credibility**
- Author profile
- AUthor belongs to the organization with greater credibility
- More publications
- More talks
- Mostly followed
